```python
import math

QuadNums = []
NeedVars = ['a','b','c']
VarNum = 0
Num = ''
ErrorMsg = ''

while VarNum < 3:

    Num = input('Enter ' + NeedVars[VarNum] +': ')
    ErrorMsg = ('You entered ' + Num + '\nPlease try again using only integers, decimal numbers and/or fractions.')           
  
    if Num.find('/') > -1:
        splitA = Num.split('/')
        if len(splitA) > 2:
            print(ErrorMsg)
            continue
        numerator = splitA[0]
        denominator = splitA[1]

        if int(denominator) == 0:
            print('denominator cannot equal 0')
            continue
        
        try:
            an = float(numerator)
            ad = float(denominator)
            a = float(an/ad)
            QuadNums.append(a)
        except:
            print(ErrorMsg)
            continue
    
    else: 
        try:
            aToFloat = float(Num)
            a = float(round(aToFloat,3))
            QuadNums.append(a)
        except: 
            print(ErrorMsg)
            continue

    if a == 0 and VarNum == 0:
        print('a cannot equal 0.')
        QuadNums = []
        continue

    VarNum+=1

a = QuadNums[0]
b = QuadNums[1]
c = QuadNums[2]

if (4*a*c) > b**2:
    print('There are no real solutions.')

# redundant with next elif, but prevents answer of -0 when b=0 and c=0
elif b**2 == 0 and 4*a*c == 0:
    print(0)

elif b**2 == 4*a*c:
    root = -b / (2*a)
    print(float(round(root,3)))

else: 
    root1 = (-b + math.sqrt(b**2 - 4*a*c))/(2*a)
    root2 = (-b - math.sqrt(b**2 - 4*a*c))/(2*a)
    print(float(round(root1,3)))
    print(float(round(root2,3)))
```
